```python
from flask import Flask, render_template, request,jsonify
from flask_cors import CORS,cross_origin
import requests
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen as uReq
import json
```

```python
uClient = uReq("https://courses.ineuron.ai/")
iNeuronPage = uClient.read()
iNeuron_html = bs(iNeuronPage,"html.parser")
uClient.close()
iNeuronHeader = iNeuron_html.find_all("body")
scriptResult=iNeuronHeader[0].find("script",{"id":"__NEXT_DATA__"})
scriptResultContent = scriptResult.contents[0]
scriptDict=dict(json.loads(scriptResultContent))
mainDict=scriptDict.get("props").get("pageProps").get("initialState").get("init").get("categories")
catDict={}
allCat = set()
for k,v in mainDict.items():
    print(mainDict[k]['title'])
    subCat = mainDict[k]['subCategories']
    for a in subCat:
        print("                        ","https://courses.ineuron.ai/category/"+subCat[a]['title'])
        courseResponse = requests.get("https://courses.ineuron.ai/category/"+subCat[a]['title'])
        courseResponse.encoding = 'utf-8'
        courseResponse_html = bs(courseResponse.text, "html.parser")
        courseResult=courseResponse_html.find_all("script",{"id":"__NEXT_DATA__"})
        courseResultContent = courseResult[0].contents
        courseDict = dict(json.loads(courseResultContent[0]))
        courseDetailDict = courseDict['props']['pageProps']['initialState']['init']['categories']
        for detKey,detVal in courseDetailDict.items():
            print(courseDetailDict[detKey]['title'])
            for detailCourseKey in courseDetailDict[detKey]['subCategories']:
                print("https://courses.ineuron.ai/"+courseDetailDict[detKey]['subCategories'][detailCourseKey]['title'])
        break
```
