```python
import pandas as pd
import json
import requests
```

```python
reqResponse = requests.get("https://api.github.com/repos/pandas-dev/pandas/issues")
response=reqResponse.json()
df = pd.DataFrame()
ddd = df.from_dict(response)
parentDf = ddd [["id","url","repository_url","comments_url"]]
userDf=df.from_dict(list(ddd['user']))
finalDataFrame=pd.concat([parentDf,userDf],axis=1)
finalDataFrame
```
